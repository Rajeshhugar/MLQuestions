## SVM

### Q71. Explain linear SVM classification.

`Linear Support Vector Machine (SVM) Classification` is a type of machine learning algorithm that is used for classification problems. It is a linear classifier that uses the concept of support vectors to find the optimal hyperplane that separates the classes in the feature space.

In SVM classification, the goal is to find a hyperplane that separates the data into two classes, with the largest margin possible. The margin is defined as the distance between the hyperplane and the closest data points from each class, also known as support vectors.

The optimization problem in SVM classification is a convex optimization problem, which can be solved using quadratic programming. The solution to this optimization problem is a hyperplane that separates the classes, with the largest margin possible.

Once the hyperplane is found, it can be used to make predictions for new data by assigning a class label to a new data point based on which side of the hyperplane it lies. If the new data point lies on one side of the hyperplane, it is assigned to the class represented by that side, and if it lies on the other side, it is assigned to the other class.

Linear SVM classification is a powerful and effective method for solving linear classification problems, especially when the data is linearly separable. It is also robust to noise and outliers, and has the ability to handle high-dimensional data. However, it may not perform well for non-linear classification problems, and in such cases, non-linear SVM methods or other classification algorithms may be more appropriate.

### Q72. What is large margin classification?

Large margin classification is a type of machine learning classification method in which the model tries to find the maximum margin between the classes in the feature space. The margin is defined as the distance between the closest data points of two different classes, which are also called support vectors. The goal of large margin classification is to maximize the margin while still correctly classifying all the training examples. This approach helps to increase the robustness and generalization performance of the model. It is commonly used in Support Vector Machines (SVM) for classification problems.

### Q73. What is a support vector?

A support vector is a data point in a dataset used for training a machine learning model, that lies closest to the decision boundary. The decision boundary is a hyperplane that separates the data points of different classes in a multi-class classification problem, or separates positive and negative instances in a binary classification problem. In the context of Support Vector Machines (SVM), support vectors play a crucial role in defining the decision boundary and determining the maximum margin between the classes. The maximum margin is the distance between the closest data points of two different classes, and the support vectors are the data points that lie on the margin or closest to the margin. Support vectors have the property that small changes in their position can result in significant changes in the position of the decision boundary, making them important in defining the model's performance and generalization ability.





### Q74. Are SVMs sensitive to scale of data?

Yes, Support Vector Machines (SVMs) can be sensitive to the scale of the data. In SVM, the distances between the data points and the decision boundary are used to determine the margin between the classes. If the scale of the data varies greatly between different features, it can result in the model placing more weight on certain features, causing the decision boundary to be affected and potentially leading to suboptimal performance. To address this issue, it is common to normalize the data by scaling each feature to have a zero mean and unit variance. This allows each feature to contribute equally to the model, regardless of its scale. Another approach is to use kernel functions in SVM, which are mathematical functions that map the original data into a higher-dimensional space, where the distances between the points are less sensitive to the scale of the data.

### Q75. What is hard margin classification? What is it's drawback?

Hard margin classification is a method used in Support Vector Machines (SVMs) to separate two classes by finding the maximum margin hyperplane that separates the classes. In this approach, the margin is defined as the distance between the hyperplane and the closest data points from either class, called support vectors. The goal of hard margin classification is to find the hyperplane that maximizes the margin, subject to the constraint that all data points are correctly classified.

The drawback of hard margin classification is that it requires the data to be linearly separable and assumes that the data is noise-free. In real-world applications, the data is often not linearly separable and may contain outliers or noise. In such cases, the hard margin classification approach can result in overfitting the data, leading to poor generalization performance on new unseen data. To address this issue, soft margin classification can be used, which allows for some data points to be misclassified, but still maximizes the margin as much as possible.





### Q76. What is soft margin classification?

Soft margin classification is a variation of the Support Vector Machine (SVM) algorithm that allows for some misclassified examples (data points) in the training set. Unlike hard margin classification, which aims to find the maximum margin hyperplane that separates the data into two classes with no misclassifications, soft margin classification allows for some degree of tolerance for misclassified examples, which is controlled by a hyperparameter "C". A small value of C results in a wider margin, allowing more misclassifications, while a larger value of C results in a narrower margin, with fewer misclassified examples. Soft margin classification is useful when the data is not linearly separable, or when there is some overlap between the classes.

### Q77. What is a kernel in SVM?

In Support Vector Machines (SVMs), a kernel is a function that transforms the input data into a higher dimensional space where it may become linearly separable. The goal of using a kernel is to find a non-linear decision boundary that separates the classes in the input data.



Kernels are used to map the original data into a higher dimensional feature space where a linear classifier can be applied. This transformation is done implicitly and allows SVMs to find the optimal decision boundary, even if the original data is not linearly separable. The most commonly used kernels in SVM include `linear, polynomial, radial basis function (RBF), and sigmoid.`

The choice of the kernel function depends on the specific problem and the structure of the data, and can have a significant impact on the performance of the SVM. In general, `linear kernels are good for linearly separable data, polynomial kernels for non-linearly separable data, and RBF kernels for data with a non-linear decision boundary`

### Q79. Explain non-linear SVM classification.

Non-linear SVM classification is a variant of SVM (Support Vector Machine) that is used to handle non-linear data. In traditional SVM, the data is assumed to be linearly separable and the decision boundary is a straight line or a hyperplane. However, in many real-world scenarios, data is not linearly separable, and a linear boundary cannot accurately classify the data. To handle such scenarios, non-linear SVM uses a technique called "kernel trick" to transform the input data into a high-dimensional feature space where a linear boundary can be applied to classify the data accurately. The most commonly used kernel functions are radial basis function (RBF), polynomial, and sigmoid. Non-linear SVM can achieve good results in a wide range of applications, including image classification, text classification, and other non-linear problems.

### Q80. What is polymonial kernel and how can we use it for non-linear SVM classification?

The polynomial kernel is a non-linear kernel function that transforms input data into a higher-dimensional space, allowing for the use of linear classifiers to solve non-linear problems. In SVM classification, a polynomial kernel can be used to model non-linear relationships between input features and target variables. The polynomial kernel is defined as:

K(x, y) = (1 + x.y)^d

Where x and y are input vectors, d is the degree of the polynomial, and x.y is the dot product of the vectors. The higher the degree of the polynomial, the more complex the non-linear decision boundary that can be modeled. To use a polynomial kernel in an SVM classifier, the "kernel" parameter in the SVM classifier model can be set to "poly". The degree of the polynomial can be specified with the "degree" parameter.





### Q81. Is there any limitation of polymonial kernel?

Choosing the right degree: Choosing the right degree for the polynomial can be challenging, as a high degree can lead to overfitting, while a low degree may not capture the non-linear relationship between the features.

May not be suitable for large datasets: Polynomial kernel can be computationally expensive and may not be suitable for large datasets or streaming data.

### Q82. What is kernel trick?

The kernel trick is a technique used in Support Vector Machines (SVMs) and other machine learning algorithms to perform non-linear classification. The basic idea behind the kernel trick is to transform the data into a higher-dimensional feature space where a linear boundary can be used to separate the classes. This transformation is performed implicitly by the kernel function, which takes two input samples and returns the dot product of the transformed samples. This allows the algorithm to perform non-linear classification without actually computing the explicit transformation of the data into a higher-dimensional space, making the algorithm more efficient and scalable.





### Q83. How can we use similarity features to tackle nonliner problem?

The kernel trick is a technique for transforming non-linearly separable data into a higher dimensional space where it becomes linearly separable. This is achieved by using a kernel function that calculates the similarity between two data points based on a chosen similarity metric, such as the radial basis function (RBF) or the polynomial kernel. ### Q84. What is Gaussian RBF kernel?

The Gaussian radial basis function (RBF) kernel is a popular choice for non-linear support vector machine (SVM) classification. It is defined as a similarity function between two points in a feature space. The RBF kernel measures the similarity between two points by calculating the exponential of the Euclidean distance between them, divided by a bandwidth parameter. This results in a smooth, bell-shaped curve, where points close to each other are considered similar and points far apart are considered dissimilar. The Gaussian RBF kernel is often used in SVM classification problems where the data is not linearly separable, as it has the ability to transform the data into a higher dimensional space where it may become linearly separable.

### Q86. How SVM uses decision function to make predictions?

Support Vector Machines (SVMs) use a decision function to make predictions. This decision function maps the input data into a higher dimensional space, where a hyperplane is used to separate the classes in a linear manner. The position of the hyperplane, also known as the maximum margin hyperplane, is determined by the support vectors, which are the closest points from each class to the hyperplane. The decision function outputs a signed distance from the input data to the hyperplane, where the sign represents the class of the input data. The data points that are closest to the hyperplane are the support vectors and play a crucial role in determining the position of the hyperplane. The prediction for a new input sample is based on which side of the hyperplane the sample lies on. If it is on the positive side of the hyperplane, the sample is classified as one class, while if it is on the negative side, it is classified as the other class.





### Q87. What is Dual Problem?

Dual problem in SVM refers to the optimization problem that is equivalent to the original optimization problem but with different constraints and objectives. This hyperplane is represented by the decision function that makes predictions based on the input data. The dual problem is often solved using numerical optimization techniques to maximize the margin between the two classes while also satisfying the constraints.

### Q89. What are online SVMs?

### Q90. Why is it important to scale the inputs when using SVMs?

Scaling the inputs is important when using Support Vector Machines (SVMs) because the SVM algorithm uses the dot product between the inputs and the model weights to make predictions. If the input features have very different ranges, the dot product will be dominated by the feature with the largest range. This can lead to numerical instability and also prevent the SVM from finding the optimal hyperplane that best separates the classes. Scaling the inputs to have a similar range can improve the numerical stability and the performance of the SVM model.

### Q96. Train an SVM classifier on the MNIST dataset. Since SVM classifiers are binaryclassifiers, you will need to use one-versus-the-rest to classify all 10 digits. You may want to tune the hyperparameters using small validation sets to speed up the process. What accuracy can you reach?

```python
from sklearn import datasets

iris = datasets.load_iris(as_frame=True)
X = iris.data[["petal length (cm)", "petal width (cm)"]].values
y = iris.target

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]
```

Now let's build and train 3 models:
* Remember that `LinearSVC` uses `loss="squared_hinge"` by default, so if we want all 3 models to produce similar results, we need to set `loss="hinge"`.
* Also, the `SVC` class uses an RBF kernel by default, so we need to set `kernel="linear"` to get similar results as the other two models.
* Lastly, the `SGDClassifier` class does not have a `C` hyperparameter, but it has another regularization hyperparameter called `alpha`, so we can tweak it to get similar results as the other two models.

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

C = 5
alpha = 0.05

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lin_clf = LinearSVC(loss="hinge", C=C, random_state=42).fit(X_scaled, y)
svc_clf = SVC(kernel="linear", C=C).fit(X_scaled, y)
sgd_clf = SGDClassifier(alpha=alpha, random_state=42).fit(X_scaled, y)
```

Let's plot the decision boundaries of these three models:

```python
def compute_decision_boundary(model):
    w = -model.coef_[0, 0] / model.coef_[0, 1]
    b = -model.intercept_[0] / model.coef_[0, 1]
    return scaler.inverse_transform([[-10, -10 * w + b], [10, 10 * w + b]])

lin_line = compute_decision_boundary(lin_clf)
svc_line = compute_decision_boundary(svc_clf)
sgd_line = compute_decision_boundary(sgd_clf)

# Plot all three decision boundaries
plt.figure(figsize=(11, 4))
plt.plot(lin_line[:, 0], lin_line[:, 1], "k:", label="LinearSVC")
plt.plot(svc_line[:, 0], svc_line[:, 1], "b--", linewidth=2, label="SVC")
plt.plot(sgd_line[:, 0], sgd_line[:, 1], "r-", label="SGDClassifier")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs") # label="Iris versicolor"
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo") # label="Iris setosa"
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend(loc="upper center")
plt.axis([0, 5.5, 0, 2])
plt.grid()

plt.show()
```

Close enough!
