#### Q36. What is the difference between error and cost function?

Error and cost function are related terms in machine learning, but they have a subtle difference.

**Error** refers to the difference between the actual value and the predicted value for a single example in the dataset. For a dataset with multiple examples, the average of the errors across all examples is often used as a measure of the overall performance of the model.

**Cost function**, on the other hand, is a mathematical function that summarizes the error across all examples in the dataset. The cost function is used to evaluate the performance of a model and to guide the optimization process. The cost function is a scalar value that represents the "cost" or "loss" associated with a particular set of model parameters. The goal is to minimize the cost function, which corresponds to finding the best-fitting model for the data.

- In summary, error is a measure of the difference between the actual and predicted values for a single example, while cost function is a measure of the overall performance of the model across all examples in the dataset.

#### Q37. What is gradient descent?

Gradient descent is an optimization algorithm used to minimize a cost function in machine learning. It is a first-order optimization method that iteratively updates the model parameters in the direction of the negative gradient of the cost function with respect to the parameters. The gradient of the cost function represents the rate of change of the cost with respect to the parameters, and the negative gradient points in the direction of the steepest decrease in the cost.

At each iteration, the parameters are updated by subtracting a fraction of the gradient from the previous values. This process is repeated until the cost function converges to a minimum or reaches a stopping criterion. The learning rate, which determines the size of the update, is a hyperparameter that must be set in advance. Gradient descent is widely used in machine learning for training models, such as linear regression, logistic regression, and neural networks.

#### Q38. What is random initialization in gradient descent?

Random initialization in gradient descent refers to the process of randomly initializing the model parameters before starting the optimization process. The purpose of random initialization is to prevent the optimization algorithm from getting stuck in a suboptimal solution due to the choice of initial values.

In gradient descent, the optimization process is sensitive to the initial values of the parameters, and the final solution can be greatly influenced by the starting point. If the parameters are initialized to the same values for all runs, the optimization algorithm may converge to the same suboptimal solution every time. To avoid this, the parameters are randomly initialized with small values drawn from a suitable distribution, such as a normal or uniform distribution.

Random initialization helps to break the symmetry of the cost function and allows the optimization algorithm to explore different regions of the parameter space. This can lead to a better solution and improved performance of the model. However, random initialization also introduces some randomness in the optimization process, and the final solution may differ between runs.

#### Q39. What is the role of learning rate?

- The learning rate is a hyperparameter in machine learning that determines the step size at which the optimization algorithm updates the model parameters. It controls the speed and direction of the updates and has a crucial role in the convergence of the optimization process.


- A high learning rate can cause the optimization algorithm to converge too quickly and potentially miss the global minimum of the cost function. On the other hand, a low learning rate can cause the optimization to converge too slowly or even get stuck in a local minimum.


- The appropriate value of the learning rate depends on the specifics of the problem and the model, and it is typically determined through trial and error or using a more sophisticated optimization algorithm, such as adaptive gradient descent.


- In summary, the learning rate plays a key role in controlling the optimization process and affects the convergence and the final solution of the model. It must be carefully chosen to ensure that the optimization algorithm reaches the optimal solution in a reasonable amount of time.

#### Q42. What is local and global minima?

In optimization, a local minimum and a global minimum refer to the values of the cost function that are minimized with respect to the model parameters.

A local minimum is a point in the parameter space where the cost function has a lower value than all its neighboring points, but it may not be the absolute minimum over the entire parameter space. In other words, it is a suboptimal solution that is not necessarily the best solution for the problem.

On the other hand, a global minimum is the point in the parameter space where the cost function has the absolute minimum value over the entire parameter space. This is the optimal solution for the problem and represents the best fit of the model to the data.



In summary, the optimization algorithm aims to find the global minimum of the cost function, which represents the best solution for the problem. However, the optimization process may get stuck in a local minimum, which is a suboptimal solution, if the learning rate is too small or the cost function has multiple local minima. To avoid getting stuck in a local minimum, it is important to choose an appropriate learning rate and initialize the model parameters with random values.

#### Q43. What is batch gradient descent?

Batch gradient descent is a gradient-based optimization algorithm for training machine learning models. It is a widely used optimization method for linear regression and other models that can be trained using gradient descent.

In batch gradient descent, the model parameters are updated after processing the entire training dataset. The algorithm computes the gradient of the cost function with respect to the parameters using the entire training set, and then updates the parameters in the direction of the negative gradient. The process is repeated for a fixed number of iterations or until the convergence criterion is met.

Batch gradient descent is simple to implement and can be easily parallelized, making it a popular choice for training large datasets. However, it can be slow and computationally expensive, especially for large datasets, as it requires processing the entire dataset at each iteration.

In summary, batch gradient descent is a widely used optimization algorithm for training machine learning models that can be trained using gradient descent. It is simple to implement and can be parallelized, but it can be slow for large datasets.

#### Q44. What is gradient ascent?

Gradient ascent is an optimization algorithm used to maximize a objective function, rather than minimize it as in gradient descent. It is used in various machine learning algorithms, such as in maximum likelihood estimation, neural network training, and others.

The basic idea of gradient ascent is to iteratively update the model parameters in the direction of the gradient of the objective function, rather than in the direction of the negative gradient as in gradient descent. At each iteration, the algorithm computes the gradient of the objective function with respect to the parameters and updates the parameters in the direction of the gradient.

The optimization process continues until the objective function reaches a maximum or a stopping criterion is met. Like gradient descent, gradient ascent can also be slow for large datasets, and the learning rate must be carefully chosen to ensure that the algorithm converges to the global maximum.

In summary, gradient ascent is an optimization algorithm used to maximize an objective function in machine learning. It works by updating the model parameters in the direction of the gradient of the objective function and can be used in various machine learning algorithms. However, it can be slow for large datasets and requires careful choice of the learning rate.

#### Q45. What is stochastic gradient descent?
stochastic - random and SGD - we are updting gradients for 1 record at a time 

Stochastic Gradient Descent (SGD) is a variant of the gradient descent optimization algorithm used in machine learning. Unlike batch gradient descent, which updates the model parameters using the average of the gradients of the cost function over the entire training set, SGD updates the parameters using the gradient of the cost function for a single randomly selected training example at each iteration.

The key advantage of SGD over batch gradient descent is that it is computationally much more efficient, especially for large datasets, as it only processes one training example at a time. Additionally, it is often faster to converge to a solution, as the parameters are updated more frequently, allowing for a faster exploration of the parameter space.

However, the major disadvantage of SGD is that the optimization process can be more unstable compared to batch gradient descent, as the updates are based on noisy gradients. To mitigate this issue, a learning rate schedule is often used, which decreases the learning rate over time, allowing the optimization to converge smoothly to a solution.

In summary, SGD is a fast optimization algorithm for machine learning that updates the model parameters using the gradient of the cost function for a single randomly selected training example at each iteration. It is computationally efficient and often faster to converge, but can be more unstable compared to batch gradient descent.

#### Q46. What is mini batch gradient descent?

Mini-batch gradient descent is a variant of gradient descent optimization that lies between batch gradient descent and stochastic gradient descent. In mini-batch gradient descent, the model parameters are updated based on the average of the gradients of the cost function over a small, randomly selected subset of the training data, rather than over the entire training set as in batch gradient descent or a single training example as in stochastic gradient descent.

The size of the mini-batch is a hyperparameter that can be tuned to balance between computational efficiency and convergence speed. A mini-batch size of 32 to 512 is commonly used in practice.

Mini-batch gradient descent offers a compromise between the speed and stability of stochastic gradient descent and the accuracy of batch gradient descent. It is computationally efficient, as it only processes a small subset of the training data at each iteration, and can converge faster than batch gradient descent, as the model parameters are updated more frequently. Additionally, the updates are based on more robust gradients than in stochastic gradient descent, which can reduce the optimization instability.


In summary, mini-batch gradient descent is a variant of gradient descent that updates the model parameters based on the average of the gradients of the cost function over a small, randomly selected subset of the training data. It offers a balance between the speed and stability of stochastic gradient descent and the accuracy of batch gradient descent, making it a popular choice in practice.

#### Q48. How learning curves avoids underfitting and overfitting problems?

Learning curves are a useful tool for avoiding underfitting and overfitting problems in machine learning. A learning curve plots the performance of a machine learning model on the training and validation data sets as a function of the size of the training data set.

In the case of underfitting, the model is not complex enough to capture the underlying patterns in the data. This is reflected in the learning curve as a low training score and a low validation score, with a large gap between the two. This indicates that the model is not able to fit the training data well and is also not able to generalize to the validation data. In this case, increasing the complexity of the model, such as adding more features or increasing the degree of a polynomial regression, can help to resolve the underfitting problem.

In the case of overfitting, the model is too complex and is fitting to the noise in the data rather than the underlying patterns. This is reflected in the learning curve as a high training score and a low validation score, with a small gap between the two. This indicates that the model is fitting the training data very well but is not able to generalize to the validation data. In this case, reducing the complexity of the model, such as removing features or reducing the degree of a polynomial regression, can help to resolve the overfitting problem.


In summary, learning curves help to avoid underfitting and overfitting problems by plotting the performance of a machine learning model on the training and validation data sets as a function of the size of the training data set. The gap between the training score and validation score, as well as the overall performance, can provide insight into whether the model is underfitting or overfitting and what steps can be taken to resolve these problems.


#### Q49. What is regularization?

Regularization is a technique used in machine learning to prevent overfitting of a model to the training data. Overfitting occurs when a model is too complex and fits to the noise in the data rather than the underlying patterns. This leads to poor generalization performance on unseen data, as the model has learned the idiosyncrasies of the training data that are not representative of the underlying patterns.

Regularization works by adding a penalty term to the objective function being optimized during training, which discourages the model from fitting to the noise in the data. The most common forms of regularization in machine learning are L1 regularization (also known as Lasso regularization) and L2 regularization (also known as Ridge regularization).

L1 regularization adds a penalty term equal to the absolute value of the magnitude of the model parameters, which encourages the model to have sparse solutions, with many of the model parameters being exactly equal to zero.

L2 regularization adds a penalty term equal to the square of the magnitude of the model parameters, which encourages the model to have smaller, but non-zero, parameters.

In summary, regularization is a technique used in machine learning to prevent overfitting of a model to the training data by adding a penalty term to the objective function that discourages the model from fitting to the noise in the data. The most common forms of regularization are L1 and L2 regularization, which encourage the model to have sparse or smaller parameters, respectively.

#### Q54. How can we use early stopping to reduce overfitting?

Early stopping is a regularization technique used to reduce overfitting in machine learning models. The idea is to monitor the performance of a model on a validation set during training and stop training when performance on the validation set starts to deteriorate. This can help prevent the model from overfitting to the training data by limiting the number of iterations and training steps.

To implement early stopping in practice, the training process is usually divided into two phases: training and validation. During training, the model is fit to the training data and the weights are updated. During validation, the model's performance on a validation set is monitored. If the performance on the validation set starts to deteriorate, training is stopped. The model's final weights are then saved and can be used to make predictions on new data.

In summary, early stopping helps reduce overfitting by monitoring performance on a validation set and stopping training when performance starts to deteriorate. This can help prevent the model from overfitting to the training data by limiting the number of iterations and training steps.





#### Q55. How is logistic regression used for classification?

Logistic regression is a statistical method that is commonly used for binary classification problems. It models the relationship between a set of input features and a binary output variable, and estimates the probability that a given input belongs to a particular class.

The logistic regression model is trained using a labeled training dataset, where the input features and binary output variable are known. This is typically done using maximum likelihood estimation.

Once the logistic regression model is trained, it can be used to make predictions on new, unseen data. The model computes the predicted probability for each input, and applies the threshold to produce a binary prediction. Logistic regression can also provide an estimate of the uncertainty in its predictions, which is useful for making informed decisions in certain applications.


Overall, logistic regression is a simple, yet powerful, method for solving binary classification problems, and is widely used in many fields, including biology, finance, and marketing.

#### Q56. Explain the use of softmax function in logistic regression.

The `softmax function` is often used in logistic regression for multiclass classification problems, where the goal is to predict one of several possible classes for a given input. The softmax function transforms the output of a logistic regression model into a set of class probabilities, where the sum of all probabilities is equal to 1.

In logistic regression, the output of the model is a linear combination of the input features and the model parameters. The `softmax function` is applied to this linear combination to produce the predicted class probabilities, given by:

**P_i = exp(z_i) / _j exp(z_j)**


where z_i is the linear combination for class i, and the sum is taken over all classes. The exponential function is used to ensure that all class probabilities are positive, and the normalization factor ensures that the sum of all probabilities is equal to 1.


The use of the softmax function in logistic regression provides several advantages. First, it allows the model to produce a probabilistic interpretation of its predictions, which can be useful for making informed decisions in certain applications. Second, the softmax function ensures that the model outputs are properly normalized, which can improve the stability and convergence of the training process. Finally, the softmax function can be used with the cross-entropy loss function to train a multiclass logistic regression model, which is a common approach in deep learning.

Overall, the `softmax function` is a crucial component of logistic regression for multiclass classification problems, and plays an important role in transforming the model's output into a set of meaningful class probabilities.

#### Q59. Explain softmax regression.

Softmax regression (also known as multinomial logistic regression or maximum entropy classifier) is a type of regression analysis used for multiclass classification problems, where the goal is to predict one of several possible classes for a given input. Softmax regression is a generalization of logistic regression, which is used for binary classification problems.

The softmax regression model takes a set of input features, and produces a set of class scores, which are then transformed into class probabilities using the softmax function. The softmax function is defined as:

P_i = exp(z_i) / _j exp(z_j)

where z_i is the score for class i, and the sum is taken over all classes. The exponential function is used to ensure that all class scores are positive, and the normalization factor ensures that the sum of all class probabilities is equal to 1.

The class scores in softmax regression are calculated using a linear combination of the input features and the model parameters, given by:

z_i = w_i^T x + b_i

where w_i is the weight vector for class i, x is the input vector, and b_i is the bias term for class i.

The training process for softmax regression involves finding the optimal model parameters that minimize a loss function, such as the cross-entropy loss. The cross-entropy loss measures the difference between the predicted class probabilities and the true class labels, and provides a way to penalize incorrect predictions. The optimization process can be performed using gradient descent or other optimization algorithms.

Overall, softmax regression is a simple and effective model for multiclass classification problems, and it is widely used in various applications, including image classification, speech recognition, and natural language processing.





#### Q60. What is cross entropy?

**Cross-entropy** is a loss function commonly used in supervised learning tasks, such as classification and regression. It measures the difference between the predicted probabilities and the actual probabilities, and is used to guide the optimization process in training a model.

The cross-entropy loss function is defined as:

`Loss = -  (actual * log(predicted))`

where actual is the ground truth label for each data sample, and predicted is the model's predicted probability for that label. The logarithm is taken to penalize incorrect predictions more harshly when they are confident (i.e. when the predicted probability is close to 1). The negative sign is used to make the loss a positive value, and the sum is taken over all data samples.

Minimizing the cross-entropy loss during training encourages the model to make predictions that are closer to the actual probabilities, and thus more accurate. The cross-entropy loss is widely used in deep learning due to its ease of computation and its ability to provide a smooth gradient, which is essential for effective **optimization**.

#### Q62. Suppose the features in your training set have very different scales. Which algorithms might suffer from this, and how? What can you do about it?

If the features in your training set have very different scales, the cost function will have the shape of an elongated bowl, so the Gradient Descent algorithms will take a long time to converge. To solve this you should scale the data before training the model. Note that the Normal Equation or SVD approach will work just fine without scaling. Moreover, regularized models may converge to a suboptimal solution if the features are not scaled: since regularization penalizes large weights, features with smaller values will tend to be ignored compared to features with larger values.

#### Q64. Do all Gradient Descent algorithms lead to the same model, provided you let them run long enough?

If the optimization problem is convex (such as Linear Regression or Logistic Regression), and assuming the learning rate is not too high, then all Gradient Descent algorithms will approach the global optimum and end up producing fairly similar models. However, unless you gradually reduce the learning rate, Stochastic GD and Mini-batch GD will never truly converge; instead, they will keep jumping back and forth around the global optimum. This means that even if you let them run for a very long time, these Gradient Descent algorithms will produce slightly different models.

However, if the training error is not going up, then your model is overfitting the training set and you should stop training.

#### Q66. Is it a good idea to stop Mini-batch Gradient Descent immediately when the validation error goes up?

Due to their random nature, neither Stochastic Gradient Descent nor Mini-batch Gradient Descent is guaranteed to make progress at every single training iteration. So if you immediately stop training when the validation error goes up, you may stop much too early, before the optimum is reached. A better option is to save the model at regular intervals; then, when it has not improved for a long time (meaning it will probably never beat the record), you can revert to the best saved model.

#### Q67. Which Gradient Descent algorithm (among those we discussed) will reach the
vicinity of the optimal solution the fastest? Which will actually converge? How
can you make the others converge as well?

Stochastic Gradient Descent has the fastest training iteration since it considers only one training instance at a time, so it is generally the first to reach the vicinity of the global optimum (or Mini-batch GD with a very small mini-batch size). However, only Batch Gradient Descent will actually converge, given enough training time. As mentioned, Stochastic GD and Mini-batch GD will bounce around the optimum, unless you gradually reduce the learning rate.

#### Q68. Suppose you are using Polynomial Regression. You plot the learning curves and you notice that there is a large gap between the training error and the validation error. What is happening? What are three ways to solve this?

If the validation error is much higher than the training error, this is likely because your model is overfitting the training set. One way to try to fix this is to reduce the polynomial degree: a model with fewer degrees of freedom is less likely to overfit. Another thing you can try is to regularize the modelfor example, by adding an  penalty (Ridge) or an  penalty (Lasso) to the cost function. This will also reduce the degrees of freedom of the model. Lastly, you can try to increase the size of the training set.

#### Q69. Suppose you are using Ridge Regression and you notice that the training error and the validation error are almost equal and fairly high. Would you say that the model suffers from high bias or high variance? Should you increase the regularization hyperparameter  or reduce it?

If both the training error and the validation error are almost equal and fairly high, the model is likely underfitting the training set, which means it has a high bias. You should try reducing the regularization hyperparameter .


#### Q70. Implement Batch Gradient Descent with early stopping for Softmax Regression (without using Scikit-Learn).

Let's start by loading the data. We will just reuse the Iris dataset we loaded earlier.

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

iris = load_iris(as_frame=True)
list(iris)
```

```python
X = iris.data[["petal length (cm)", "petal width (cm)"]].values
y = iris["target"].values
```

We need to add the bias term for every instance ($x_0 = 1$). The easiest option to do this would be to use Scikit-Learn's `add_dummy_feature()` function, but the point of this exercise is to get a better understanding of the algorithms by implementing them manually. So here is one possible implementation:

```python
X_with_bias = np.c_[np.ones(len(X)), X]
```

The easiest option to split the dataset into a training set, a validation set and a test set would be to use Scikit-Learn's `train_test_split()` function, but again, we want to do it manually:

```python
test_ratio = 0.2
validation_ratio = 0.2
total_size = len(X_with_bias)

test_size = int(total_size * test_ratio)
validation_size = int(total_size * validation_ratio)
train_size = total_size - test_size - validation_size

np.random.seed(42)
rnd_indices = np.random.permutation(total_size)

X_train = X_with_bias[rnd_indices[:train_size]]
y_train = y[rnd_indices[:train_size]]
X_valid = X_with_bias[rnd_indices[train_size:-test_size]]
y_valid = y[rnd_indices[train_size:-test_size]]
X_test = X_with_bias[rnd_indices[-test_size:]]
y_test = y[rnd_indices[-test_size:]]
```

The targets are currently class indices (0, 1 or 2), but we need target class probabilities to train the Softmax Regression model. Each instance will have target class probabilities equal to 0.0 for all classes except for the target class which will have a probability of 1.0 (in other words, the vector of class probabilities for any given instance is a one-hot vector). Let's write a small function to convert the vector of class indices into a matrix containing a one-hot vector for each instance. To understand this code, you need to know that `np.diag(np.ones(n))` creates an nn matrix full of 0s except for 1s on the main diagonal. Moreover, if `a` is a NumPy array, then `a[[1, 3, 2]]` returns an array with 3 rows equal to `a[1]`, `a[3]` and `a[2]` (this is [advanced NumPy indexing](https://numpy.org/doc/stable/user/basics.indexing.html#advanced-indexing)).

```python
def to_one_hot(y):
    return np.diag(np.ones(y.max() + 1))[y]
```

Let's test this function on the first 10 instances:

```python
y_train[:10]
```

```python
to_one_hot(y_train[:10])
```

Looks good, so let's create the target class probabilities matrix for the training set and the test set:

```python
Y_train_one_hot = to_one_hot(y_train)
Y_valid_one_hot = to_one_hot(y_valid)
Y_test_one_hot = to_one_hot(y_test)
```

Now let's scale the inputs. We compute the mean and standard deviation of each feature on the training set (except for the bias feature), then we center and scale each feature in the training set, the validation set, and the test set:

```python
mean = X_train[:, 1:].mean(axis=0)
std = X_train[:, 1:].std(axis=0)
X_train[:, 1:] = (X_train[:, 1:] - mean) / std
X_valid[:, 1:] = (X_valid[:, 1:] - mean) / std
X_test[:, 1:] = (X_test[:, 1:] - mean) / std
```

Now let's implement the Softmax function. Recall that it is defined by the following equation:

$\sigma\left(\mathbf{s}(\mathbf{x})\right)_k = \dfrac{\exp\left(s_k(\mathbf{x})\right)}{\sum\limits_{j=1}^{K}{\exp\left(s_j(\mathbf{x})\right)}}$

```python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = exps.sum(axis=1, keepdims=True)
    return exps / exp_sums
```

We are almost ready to start training. Let's define the number of inputs and outputs:

```python
n_inputs = X_train.shape[1]  # == 3 (2 features plus the bias term)
n_outputs = len(np.unique(y_train))  # == 3 (there are 3 iris classes)
```

Now here comes the hardest part: training! Theoretically, it's simple: it's just a matter of translating the math equations into Python code. But in practice, it can be quite tricky: in particular, it's easy to mix up the order of the terms, or the indices. You can even end up with code that looks like it's working but is actually not computing exactly the right thing. When unsure, you should write down the shape of each term in the equation and make sure the corresponding terms in your code match closely. It can also help to evaluate each term independently and print them out. The good news it that you won't have to do this everyday, since all this is well implemented by Scikit-Learn, but it will help you understand what's going on under the hood.

So the equations we will need are the cost function:

$J(\mathbf{\Theta}) =
- \dfrac{1}{m}\sum\limits_{i=1}^{m}\sum\limits_{k=1}^{K}{y_k^{(i)}\log\left(\hat{p}_k^{(i)}\right)}$

And the equation for the gradients:

$\nabla_{\mathbf{\theta}^{(k)}} \, J(\mathbf{\Theta}) = \dfrac{1}{m} \sum\limits_{i=1}^{m}{ \left ( \hat{p}^{(i)}_k - y_k^{(i)} \right ) \mathbf{x}^{(i)}}$

Note that $\log\left(\hat{p}_k^{(i)}\right)$ may not be computable if $\hat{p}_k^{(i)} = 0$. So we will add a tiny value $\epsilon$ to $\log\left(\hat{p}_k^{(i)}\right)$ to avoid getting `nan` values.

```python
eta = 0.5
n_epochs = 5001
m = len(X_train)
epsilon = 1e-5

np.random.seed(42)
Theta = np.random.randn(n_inputs, n_outputs)

for epoch in range(n_epochs):
    logits = X_train @ Theta
    Y_proba = softmax(logits)
    if epoch % 1000 == 0:
        Y_proba_valid = softmax(X_valid @ Theta)
        xentropy_losses = -(Y_valid_one_hot * np.log(Y_proba_valid + epsilon))
        print(epoch, xentropy_losses.sum(axis=1).mean())
    error = Y_proba - Y_train_one_hot
    gradients = 1 / m * X_train.T @ error
    Theta = Theta - eta * gradients
```

And that's it! The Softmax model is trained. Let's look at the model parameters:

```python
Theta
```

Let's make predictions for the validation set and check the accuracy score:

```python
logits = X_valid @ Theta
Y_proba = softmax(logits)
y_predict = Y_proba.argmax(axis=1)

accuracy_score = (y_predict == y_valid).mean()
accuracy_score
```

Well, this model looks pretty ok. For the sake of the exercise, let's add a bit of $\ell_2$ regularization. The following training code is similar to the one above, but the loss now has an additional $\ell_2$ penalty, and the gradients have the proper additional term (note that we don't regularize the first element of `Theta` since this corresponds to the bias term). Also, let's try increasing the learning rate `eta`.

```python
eta = 0.5
n_epochs = 5001
m = len(X_train)
epsilon = 1e-5
alpha = 0.01  # regularization hyperparameter

np.random.seed(42)
Theta = np.random.randn(n_inputs, n_outputs)

for epoch in range(n_epochs):
    logits = X_train @ Theta
    Y_proba = softmax(logits)
    if epoch % 1000 == 0:
        Y_proba_valid = softmax(X_valid @ Theta)
        xentropy_losses = -(Y_valid_one_hot * np.log(Y_proba_valid + epsilon))
        l2_loss = 1 / 2 * (Theta[1:] ** 2).sum()
        total_loss = xentropy_losses.sum(axis=1).mean() + alpha * l2_loss
        print(epoch, total_loss.round(4))
    error = Y_proba - Y_train_one_hot
    gradients = 1 / m * X_train.T @ error
    gradients += np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients
```

Because of the additional $\ell_2$ penalty, the loss seems greater than earlier, but perhaps this model will perform better? Let's find out:

```python
logits = X_valid @ Theta
Y_proba = softmax(logits)
y_predict = Y_proba.argmax(axis=1)

accuracy_score = (y_predict == y_valid).mean()
accuracy_score
```

In this case, the $\ell_2$ penalty did not change the test accuracy. Perhaps try fine-tuning `alpha`?

Now let's add early stopping. For this we just need to measure the loss on the validation set at every iteration and stop when the error starts growing.

```python
eta = 0.5
n_epochs = 50_001
m = len(X_train)
epsilon = 1e-5
C = 100  # regularization hyperparameter
best_loss = np.infty

np.random.seed(42)
Theta = np.random.randn(n_inputs, n_outputs)

for epoch in range(n_epochs):
    logits = X_train @ Theta
    Y_proba = softmax(logits)
    Y_proba_valid = softmax(X_valid @ Theta)
    xentropy_losses = -(Y_valid_one_hot * np.log(Y_proba_valid + epsilon))
    l2_loss = 1 / 2 * (Theta[1:] ** 2).sum()
    total_loss = xentropy_losses.sum(axis=1).mean() + 1 / C * l2_loss
    if epoch % 1000 == 0:
        print(epoch, total_loss.round(4))
    if total_loss < best_loss:
        best_loss = total_loss
    else:
        print(epoch - 1, best_loss.round(4))
        print(epoch, total_loss.round(4), "early stopping!")
        break
    error = Y_proba - Y_train_one_hot
    gradients = 1 / m * X_train.T @ error
    gradients += np.r_[np.zeros([1, n_outputs]), 1 / C * Theta[1:]]
    Theta = Theta - eta * gradients
```

```python
logits = X_valid @ Theta
Y_proba = softmax(logits)
y_predict = Y_proba.argmax(axis=1)

accuracy_score = (y_predict == y_valid).mean()
accuracy_score
```

Oh well, still no change in validation accuracy, but at least early stopping shortened training a bit.

Now let's plot the model's predictions on the whole dataset (remember to scale all features fed to the model):

```python
custom_cmap = mpl.colors.ListedColormap(['#fafab0', '#9898ff', '#a0faa0'])

x0, x1 = np.meshgrid(np.linspace(0, 8, 500).reshape(-1, 1),
                     np.linspace(0, 3.5, 200).reshape(-1, 1))
X_new = np.c_[x0.ravel(), x1.ravel()]
X_new = (X_new - mean) / std
X_new_with_bias = np.c_[np.ones(len(X_new)), X_new]

logits = X_new_with_bias @ Theta
Y_proba = softmax(logits)
y_predict = Y_proba.argmax(axis=1)

zz1 = Y_proba[:, 1].reshape(x0.shape)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.plot(X[y == 2, 0], X[y == 2, 1], "g^", label="Iris virginica")
plt.plot(X[y == 1, 0], X[y == 1, 1], "bs", label="Iris versicolor")
plt.plot(X[y == 0, 0], X[y == 0, 1], "yo", label="Iris setosa")

plt.contourf(x0, x1, zz, cmap=custom_cmap)
contour = plt.contour(x0, x1, zz1, cmap="hot")
plt.clabel(contour, inline=1)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend(loc="upper left")
plt.axis([0, 7, 0, 3.5])
plt.grid()
plt.show()
```

And now let's measure the final model's accuracy on the test set:

```python
logits = X_test @ Theta
Y_proba = softmax(logits)
y_predict = Y_proba.argmax(axis=1)

accuracy_score = (y_predict == y_test).mean()
accuracy_score
```

Well we get even better performance on the test set. This variability is likely due to the very small size of the dataset: depending on how you sample the training set, validation set and the test set, you can get quite different results. Try changing the random seed and running the code again a few times, you will see that the results will vary.
